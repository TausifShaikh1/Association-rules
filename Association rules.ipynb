```python
import pandas as pd
df = pd.read_csv('/content/groceries.csv')
df
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
groceries = []
```

```python
with open('/content/groceries.csv') as f:
    groceries = f.read()
```

```python
groceries
```

```python
groceries = groceries.split("\n")
groceries
```

```python
groceries_list = []
```

```python
for i in groceries:
  groceries_list.append(i.split(','))
```

```python
groceries_list
```

```python
all_groceries_list = [i for item in groceries_list for i in item ]
```

```python
all_groceries_list
```

```python
from collections import Counter
```

```python
item_frequencies = Counter(all_groceries_list)
```

```python
item_frequencies
```

```python
item_frequencies = sorted(item_frequencies.items(), key = lambda x:x[1])
```

```python
item_frequencies
```

```python
frequencies = list(reversed([i[1] for i in item_frequencies]))
```

```python
frequencies
```

```python
items = list(reversed([i[0] for i in item_frequencies]))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(height = frequencies[1:11], x = list(range(1, 11)), color = 'red')
plt.xticks(list(range(1, 11), ), items[1:11], rotation = 80)
plt.xlabel("items")
plt.ylabel("Count")
plt.show
```

```python
groceries_series = pd.DataFrame(pd.Series(groceries_list))
groceries_series
```

```python
groceries_series = groceries_series.iloc[:9835, :]
```

```python
groceries_series.columns = ["transactions"]
```

```python
groceries_series
```

```python
X = groceries_series['transactions'].str.join(sep = '*').str.get_dummies(sep = '*')
X
```

```python
frequent_itemsets = apriori(X, min_support= 0.0075, max_len= 4, use_colnames= True)
```

```python
frequent_itemsets
```

```python
frequent_itemsets.sort_values('support', ascending= False, inplace= True)
frequent_itemsets
```

```python
plt.bar(x = list(range(0, 11)), height= frequent_itemsets.support[0:11], color = 'blue')
plt.xticks(list(range(0, 11)), frequent_itemsets.itemsets[0:11], rotation = 80)
plt.xlabel('itemsets')
plt.ylabel('support')
plt.show
```

```python
rules = association_rules(frequent_itemsets, metric = "lift", min_threshold= 1)
rules
```
